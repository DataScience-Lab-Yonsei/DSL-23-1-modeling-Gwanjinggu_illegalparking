```python
df = df.transpose()
df.columns = ['수요']
df
```

```python
df['수요'] = [25259, 6199, 5123, 4181, 4266, 5304, 3560, 7413, 2283, 1074, -2390, 8391, 1789, -3481, 11405]
df
```

```python
df.reset_index(inplace=True)
df
```

```python
df.columns = ['동', '수요']
df
```

```python
# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/vuski/admdongkor/master/ver20230101/HangJeongDong_ver20230101.geojson')
c = r.content
seoul_geo = json.loads(c)
```

```python
seoul_geo = seoul_geo['features'][65:80]
```

```python
#geojson 형식 맞추기
geo = {"type": "FeatureCollection",
  "features": seoul_geo}
geo
```

```python
seoul = folium.Map(location=[37.559819, 126.963895], 
                   zoom_start = 11,
                   tiles='cartodbpositron'
                   )
```

```python
folium.GeoJson(
    geo,
    style_function = lambda x: {'fillColor':'#00000000', 'color':'black', 'weight' :'2'}
).add_to(seoul)

seoul
```

```python
folium.Choropleth(
    geo_data=geo, #행정구역 경계 JSON; 구역을 나누고
    data=df, #거기에 맞는 지표
    columns = ['동', '수요'], #df에서 사용할 것들
    key_on = 'properties.temp', #df와 일치시킬 JSON의 이름
    fill_color = 'Blues',
    legend_name ='주차수요',
).add_to(seoul)
seoul

#노란색: 'YlOrRd'
```
